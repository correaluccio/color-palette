# Obtención de la paleta de colores de una imagen con K-means clustering.

#### Con la funcion get_pixels se obtiene un dataframe con los datos de los pixeles de la imagen

```

from PIL import Image
import pandas as pd

def get_pixels(image_patch,res):
    pixels = {"r": [], "g":[], "b":[]}
    pixels_df = pd.DataFrame(pixels)
    img = Image.open(image_patch)
    ancho,alto = img.size
    for x in range(0,ancho,res):
        for y in range(0,alto,res):
            r,g,b = img.getpixel((x,y))
            new_pixel = {"r":[r/256],"g":[g/256],"b":[b/256]}
            new_pixel_df = pd.DataFrame(new_pixel)
            pixels_df = pd.concat([pixels_df,new_pixel_df],ignore_index = True)
    return pixels_df
    
```

Los parámetros de entrada son image_patch que hace referencia a la imagen a la que queremos extraerle la paleta de colores principal y res que divide el ancho y el alto de la imagen para obtener menos cantidad de pixels, ya que, por ejemplo, en una imagen de 1920*1080 el dataframe tendría 2073600 filas, se recomienda un valor de res entre 10 y 100, dependiendo de la resolución de la imagen.


### Para ilustrar el método se utilizará la siguiente imagen

![cataratas_iguazu](cataratas.JPG)

Los pixeles asociados a esta foto son:

````python
cataratas = get_pixels("cataratas.JPG",50)
````

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cataratas['r'], cataratas['g'], cataratas['b'], c='r', marker='o')
ax.set_xlabel('r')
ax.set_ylabel('g')
ax.set_zlabel('b')
````

````python
import numpy as np
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
````

````python
kmeans = KMeans(n_clusters=5).fit(cataratas)
centroids = kmeans.cluster_centers_
print(centroids)
````

````python
labels = kmeans.predict(cataratas)
cataratas["labels"] = labels
````

````python
cataratas
````

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cataratas['r'], cataratas['g'], cataratas['b'], c=centroids[cataratas["labels"]], marker='o')
ax.set_xlabel('r')
ax.set_ylabel('g')
ax.set_zlabel('b')
````
